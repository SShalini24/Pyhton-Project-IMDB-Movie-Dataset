**Classify Movies Based on Rating (Excellent, Good and Average)**

```python
data['Rating'].unique()
```

```python
def rating(rating):
  if rating >=7.0:
    return "Excellent"
  elif rating >=6.0:
    return "Good"
  else:
    return "Average"
```

```python
data['rating_cat'] = data['Rating'].apply(rating)
```

```python
data['rating_cat']
```

**Find Unique Values from the Genre**

```python
list1 = []
for value in data['Genre']:
  list1.append(value.split(","))
```

```python
list1
```

```python
one_d = []
for item in list1:
  for item1 in item:
    one_d.append(item1)
```

```python
one_d
```

```python
uni_list = []
for item in one_d:
  if item not in uni_list:
    uni_list.append(item)
```

```python
len(uni_list)
```

```python
# above are the unique values in the genre
```
